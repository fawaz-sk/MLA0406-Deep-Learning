```python
from sklearn.tree import DecisionTreeClassifier

# Training data
X = [[25], [35], [45], [20], [50], [30]]
y = ["No", "Yes", "Yes", "No", "Yes", "No"]

# Create and train model
model = DecisionTreeClassifier()
model.fit(X, y)

# User input
age = int(input("Enter Age: "))

# Prediction
result = model.predict([[age]])

print("Loan Approval:", result[0])
```
