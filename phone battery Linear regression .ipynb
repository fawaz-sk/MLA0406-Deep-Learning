```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Training data
screen_hours = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
battery_left = np.array([95, 85, 75, 65, 55, 45])

# Create and train model
model = LinearRegression()
model.fit(screen_hours, battery_left)

# User input
hours = float(input("Enter screen usage hours: "))

# Predict battery percentage
prediction = model.predict([[hours]])

print("Predicted Battery Remaining: {:.2f}%".format(prediction[0]))
```
